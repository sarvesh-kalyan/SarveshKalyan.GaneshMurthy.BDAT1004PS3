##### Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt

data_mar_div = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data_mar_div.head()
```

```python
plt.figure()

data_mar_div.plot(x='Year',y=['Marriages_per_1000','Divorces_per_1000'],label=['Marriage','Divorce'])

plt.title('Marriages & Divorces', color = '#C06C84', fontsize = 15)

plt.xlabel('Year', color = '#F8B195')
plt.ylabel('Capita', color = '#F67280')

plt.show()
```
